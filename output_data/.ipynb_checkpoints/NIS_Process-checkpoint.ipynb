```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
nis_radar = pd.read_csv('./NISRec_radar.txt', header=None)
%matplotlib inline
plt.plot(nis_radar)
plt.hlines(7.81, 0, len(nis_radar), 'r', label='95%')
plt.xlabel('Measurements')
plt.ylabel('Radar NIS')
plt.title('Radar Consistency Check')
```

```python
nis_laser = pd.read_csv('./NISRec_laser.txt', header=None)
plt.plot(nis_laser)
plt.hlines(5.99, 0, len(nis_radar), 'r', label='95%')
plt.xlabel('Measurements')
plt.ylabel('Laser NIS')
```
